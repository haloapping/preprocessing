# Content

##### Price Tomato
Tomato Price across markets in karnataka from 2016-2018

##### Features
- Market: name of the market where vegetable sold
- Arrival Date: date Vegetable arrived
- Arrivals (Tonnes): number of tonnes arrives
- Variety: variety of the tomato
- Minimum Price(Rs./Quintal): minimum Price Per Quintal(100 kgs)
- Maximum Price(Rs./Quintal): maximum Price Per Quintal(100kgs)
- Modal Price(Rs./Quintal): Average Price
