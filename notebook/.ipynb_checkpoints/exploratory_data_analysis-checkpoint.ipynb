# Import modules

```python
import pandas as pd
import plotly.express as px
```

***
# Initialize and load dataset

```python
raw_df = pd.read_csv('../dataset/raw_no_mv_dataset.csv')
```

```python
raw_df.head()
```

# Univariate Analysis

```python
raw_df.describe(include='all')
```

# Graphical Univariate Analysis

### - market

```python
fig = px.histogram(raw_df, y="market", nbins=40)

fig.update_layout(
    title   = 'Market Histogram',
    title_x = 0.5
)

fig.show()
```

### - arrival_date

```python
fig = px.histogram(raw_df, x="arrival_date", nbins=844)

fig.update_layout(
    title   = 'Arrival Date Histogram',
    title_x = 0.5
)

fig.show()
```

### - arrivals_tonnes

```python
fig = px.histogram(raw_df, x="arrivals_tonnes", nbins=2000)

fig.update_layout(
    title   = 'Arrival Tonnes Histogram',
    title_x = 0.5
)

fig.show()
```

### - variety

```python
fig = px.histogram(raw_df, x="variety")

fig.update_layout(
    title   = 'Variety Histogram',
    title_x = 0.5,
    height  = 500,
    width   = 500
)

fig.show()
```

### - minimum_price

```python
fig = px.histogram(raw_df, x="minimum_price", nbins=679)

fig.update_layout(
    title   = 'Minimum Price Histogram',
    title_x = 0.5,
)

fig.show()
```

### - maximum_price

```python
fig = px.histogram(raw_df, x="maximum_price", nbins=932)

fig.update_layout(
    title   = 'Maximum Price Histogram',
    title_x = 0.5,
)

fig.show()
```
